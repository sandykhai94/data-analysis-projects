```python
# create a series called s_dictionary from the dictionary: {"brand": "Ford", "model": "Mustang", "year": 1964}

import pandas as pd
car_data = {"brand": "Ford", "model": "Mustang", "year": 1964}
car_series = pd.Series(car_data, name="Car Specifications")
print(car_series)
```

```python
import pandas as pd
height_values = [5.3, 6.2, 5.8, 5.0, 5.5]
weight_values = [65, 89, 75, 60, 59]
indexes = ['Person 1', 'Person 2', 'Person 3', 'Person 4', 'Person 5']
height = pd.Series(data=height_values, index=indexes)
weight = pd.Series(data=weight_values, index=indexes)
person_data = pd.DataFrame({
    'Height (feet)': height,
    'Weight (kg)': weight
})
print("Person Data DataFrame:")
print(person_data)
```

```python
import pandas as pd
height_values = [5.3, 6.2, 5.8, 5.0, 5.5]
weight_values = [65, 89, 75, 60, 59]
indexes = ['Person 1', 'Person 2', 'Person 3', 'Person 4', 'Person 5']
height = pd.Series(data=height_values, index=indexes)
weight = pd.Series(data=weight_values, index=indexes)
data_dict = {
    "height": height,
    "weight": weight
}
df = pd.DataFrame(data_dict)
print("DataFrame from height and weight series:")
print(df)
```
