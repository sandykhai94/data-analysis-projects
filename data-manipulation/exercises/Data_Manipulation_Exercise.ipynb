```python
### Using a function to create a new column 
# Write your max age check here:

max_age = data['age'].max()
print("Maximum age:", max_age)
bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]  
labels = ["0-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80-89", "90-99"]
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, include_lowest=True)
data.head()
```
